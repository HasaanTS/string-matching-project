```python
import requests
from bs4 import BeautifulSoup
import collections
import pandas as pd
import numpy as np
import os
```

```python
def file_write(df, path_and_name, idx=False):
    #This is a simple function that writes a dataframe to a csv file
    return df.to_csv(path_and_name+'.csv', sep = ';', index = idx)#Making index False helps ignore index column when writing file
```

```python
def download_file(url):
    local_filename = str(url.split('?')[1])+'.csv'
    # NOTE the stream=True parameter below
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192): 
                if chunk: # filter out keep-alive new chunks
                    f.write(chunk)
                    # f.flush()
    return local_filename
```

```python
def groupDownload(path, URL):
    _website = 'https://www.opensecrets.org'
    page = requests.get(URL)
    soup = BeautifulSoup(page.content)
    _options = soup.find_all('option')
    for _option in _options:
        _link = str(_option['value'])
        _link = _link.split('?')
        _link = str(_link[0])+'.csv?'+str(_link[1])
        os.chdir(path)
        download_file(url=_website+_link)
```
