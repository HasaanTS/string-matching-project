```python
import requests
def download_file(url):
#     local_filename = url.split('/')[-1]
    local_filename = str(url.split('?')[1])+'.csv'
    # NOTE the stream=True parameter below
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192): 
                if chunk: # filter out keep-alive new chunks
                    f.write(chunk)
                    # f.flush()
    return local_filename
```

```python
nameLinkDB = collections.OrderedDict()
personName_list = dataTable.find_all('a', href=True)
# personName_list = soup.find_all('a', href=True)
for personName in personName_list:
#     print(personName.prettify())
    name = personName.contents[0]
    print name
    links = siteStr + personName.get('href')
    print links
    nameLinkDB[name] = links
```

```python
# _str = 'https://www.opensecrets.org/personal-finances/net-worth?cid=N00035616&year=2015'
def stripper(_str):
    _str = _str.split('?')[1]
#     print _str
    _str = _str.split('&')
#     print _str
    ID = _str[0].split('=')[1]
    YEAR = _str[1].split('=')[1]
    return ID, YEAR
# stripper(_str)
person_ID = []
person_YEAR = []
for value in nameLinkDB.values():
    ID, YEAR = stripper(value)
    person_ID.append(ID)
    person_YEAR.append(YEAR)
```

```python
nameLinkDF = pd.DataFrame({'name':nameLinkDB.keys(), 'link':nameLinkDB.values(), 'id':person_ID, 'year':person_YEAR}, index=np.arange(len(nameLinkDB)))
```
